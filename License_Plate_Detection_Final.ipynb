```python
#Enhancing video code (if needed)
import cv2
import numpy as np

input_path = 'cars.mp4'
output_path = 'cars_enhanced.mp4'

cap = cv2.VideoCapture(input_path)

fps = int(cap.get(cv2.CAP_PROP_FPS))
frame_width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

def enhance_frame(frame):
    lab = cv2.cvtColor(frame, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    l = clahe.apply(l)
    enhanced_lab = cv2.merge((l, a, b))
    frame = cv2.cvtColor(enhanced_lab, cv2.COLOR_LAB2BGR)

    kernel = np.array([[0, -1, 0],
                       [-1, 5,-1],
                       [0, -1, 0]])
    frame = cv2.filter2D(frame, -1, kernel)

    return frame

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    enhanced_frame = enhance_frame(frame)
    out.write(enhanced_frame)

cap.release()
out.release()
cv2.destroyAllWindows()
print("✅ Enhanced video saved:", output_path)
```

```python
#code for function
import string
import easyocr

READER = easyocr.Reader(['en'], gpu=False)

def write_csv(results, output_path):

    header = ['frame_nmr', 'car_id', 'car_bbox',
              'license_plate_bbox', 'license_plate_bbox_score',
              'license_number', 'license_number_score']

    with open(output_path, 'w') as f:
        f.write(','.join(header) + '\n')

        for frame, cars in results.items():
            for car_id, data in cars.items():
                if all(k in data for k in ['car', 'license_plate']) and 'text' in data['license_plate']:
                    car_bbox = '[{} {} {} {}]'.format(*data['car']['bbox'])
                    lp_bbox = '[{} {} {} {}]'.format(*data['license_plate']['bbox'])
                    row = [
                        frame,
                        car_id,
                        car_bbox,
                        lp_bbox,
                        data['license_plate']['bbox_score'],
                        data['license_plate']['text'],
                        data['license_plate']['text_score']
                    ]
                    f.write(','.join(map(str, row)) + '\n')

CHAR_TO_INT = {
    'O': '0', 'I': '1', 'J': '3',
    'A': '4', 'G': '6', 'S': '5'
}

INT_TO_CHAR = {
    '0': 'O', '1': 'I', '3': 'J',
    '4': 'A', '6': 'G', '5': 'S'
}

def license_complies_format(text):
    if len(text) != 7:
        return False

    valid_alpha = string.ascii_uppercase + ''.join(INT_TO_CHAR.keys())
    valid_numeric = '0123456789' + ''.join(CHAR_TO_INT.keys())

    return (
        text[0] in valid_alpha and
        text[1] in valid_alpha and
        text[2] in valid_numeric and
        text[3] in valid_numeric and
        text[4] in valid_alpha and
        text[5] in valid_alpha and
        text[6] in valid_alpha
    )


def format_license(text):
    formatted = ''
    position_map = {
        0: INT_TO_CHAR, 1: INT_TO_CHAR, 2: CHAR_TO_INT,3: CHAR_TO_INT,
         4: INT_TO_CHAR,5: INT_TO_CHAR, 6: INT_TO_CHAR

    }

    for i in range(7):
        char = text[i]
        if char in position_map.get(i, {}):
            formatted += position_map[i][char]
        else:
            formatted += char
    return formatted

def read_license_plate(license_plate_crop):

    detections = READER.readtext(license_plate_crop)

    for bbox, text, score in detections:
        cleaned_text = text.upper().replace(' ', '')
        if license_complies_format(cleaned_text):
            return format_license(cleaned_text), score

    return None, None

def get_car(license_plate, vehicle_track_ids):

    x1, y1, x2, y2, _, _ = license_plate

    for xcar1, ycar1, xcar2, ycar2, car_id in vehicle_track_ids:
        if x1 > xcar1 and y1 > ycar1 and x2 < xcar2 and y2 < ycar2:
            return xcar1, ycar1, xcar2, ycar2, car_id

    return -1, -1, -1, -1, -1
```

```python
#main code
import cv2
import numpy as np
from ultralytics import YOLO

VIDEO_PATH = 'cars.mp4'
OUTPUT_CSV = './test_results.csv'
VEHICLE_CLASSES = [2, 3, 5, 7]  # Car, motorcycle, bus, truck
COCO_MODEL_PATH = 'yolov8n.pt'
LICENSE_MODEL_PATH = '/content/number_plate_best.pt'

def initialize_models():
    vehicle_detector = YOLO(COCO_MODEL_PATH)
    plate_detector = YOLO(LICENSE_MODEL_PATH)
    return vehicle_detector, plate_detector


def process_frame(frame, frame_number, vehicle_detector, plate_detector, mot_tracker, results):
    """
    Process a single video frame:
    - Detect vehicles
    - Track them
    - Detect license plates
    - Match them to vehicles
    - Run OCR and save results
    """
    results[frame_number] = {}

    vehicle_detections = vehicle_detector(frame)[0]
    filtered_detections = []
    for det in vehicle_detections.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = det
        if int(class_id) in VEHICLE_CLASSES:
            filtered_detections.append([x1, y1, x2, y2, score])

    tracking_ids = mot_tracker.update(np.asarray(filtered_detections))

    license_detections = plate_detector(frame)[0]
    for plate in license_detections.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = plate

        xcar1, ycar1, xcar2, ycar2, car_id = get_car(plate, tracking_ids)
        if car_id == -1:
            continue

        license_crop = frame[int(y1):int(y2), int(x1):int(x2)]
        gray_crop = cv2.cvtColor(license_crop, cv2.COLOR_BGR2GRAY)
        _, binarized_crop = cv2.threshold(gray_crop, 64, 255, cv2.THRESH_BINARY_INV)

        text, score_text = read_license_plate(binarized_crop)

        if text:
            results[frame_number][car_id] = {
                'car': {'bbox': [xcar1, ycar1, xcar2, ycar2]},
                'license_plate': {
                    'bbox': [x1, y1, x2, y2],
                    'bbox_score': score,
                    'text': text,
                    'text_score': score_text
                }
            }


def run_license_plate_pipeline():

    cap = cv2.VideoCapture(VIDEO_PATH)
    vehicle_detector, plate_detector = initialize_models()
    mot_tracker = Sort()

    results = {}
    frame_number = -1
    ret = True

    while ret:
        frame_number += 1
        ret, frame = cap.read()
        if not ret:
            break

        process_frame(frame, frame_number, vehicle_detector, plate_detector, mot_tracker, results)

    write_csv(results, OUTPUT_CSV)

if __name__ == "__main__":
    run_license_plate_pipeline()
```

```python
# Handling Missing License Plates

import csv
import numpy as np
from scipy.interpolate import interp1d

def interpolate_missing_boxes(records):
    frames = np.array([int(entry['frame_nmr']) for entry in records])
    car_ids = np.array([int(float(entry['car_id'])) for entry in records])

    car_boxes = np.array([
        list(map(float, entry['car_bbox'][1:-1].split())) for entry in records
    ])
    plate_boxes = np.array([
        list(map(float, entry['license_plate_bbox'][1:-1].split())) for entry in records
    ])

    interpolated_output = []
    unique_ids = np.unique(car_ids)

    for cid in unique_ids:
        existing_frames = [rec['frame_nmr'] for rec in records if int(float(rec['car_id'])) == cid]
        print(existing_frames, cid)

        cid_mask = (car_ids == cid)
        cid_frames = frames[cid_mask]

        interpolated_car_boxes = []
        interpolated_plate_boxes = []

        start_frame = cid_frames[0]

        for idx in range(len(car_boxes[cid_mask])):
            current_frame = cid_frames[idx]
            current_car_box = car_boxes[cid_mask][idx]
            current_plate_box = plate_boxes[cid_mask][idx]

            if idx > 0:
                prev_frame = cid_frames[idx - 1]
                prev_car_box = interpolated_car_boxes[-1]
                prev_plate_box = interpolated_plate_boxes[-1]

                if current_frame - prev_frame > 1:
                    gap = current_frame - prev_frame
                    x_vals = np.array([prev_frame, current_frame])
                    interp_frames = np.linspace(prev_frame, current_frame, num=gap, endpoint=False)

                    car_interp = interp1d(x_vals, np.vstack((prev_car_box, current_car_box)), axis=0)
                    new_car_boxes = car_interp(interp_frames)

                    plate_interp = interp1d(x_vals, np.vstack((prev_plate_box, current_plate_box)), axis=0)
                    new_plate_boxes = plate_interp(interp_frames)

                    interpolated_car_boxes.extend(new_car_boxes[1:])
                    interpolated_plate_boxes.extend(new_plate_boxes[1:])

            interpolated_car_boxes.append(current_car_box)
            interpolated_plate_boxes.append(current_plate_box)

        for i, (car_box, plate_box) in enumerate(zip(interpolated_car_boxes, interpolated_plate_boxes)):
            frame_id = start_frame + i
            interpolated_row = {
                'frame_nmr': str(frame_id),
                'car_id': str(cid),
                'car_bbox': ' '.join(map(str, car_box)),
                'license_plate_bbox': ' '.join(map(str, plate_box))
            }

            if str(frame_id) not in existing_frames:
                interpolated_row.update({
                    'license_plate_bbox_score': '0',
                    'license_number': '0',
                    'license_number_score': '0'
                })
            else:
                matching_record = [
                    entry for entry in records
                    if int(entry['frame_nmr']) == frame_id and int(float(entry['car_id'])) == cid
                ][0]

                interpolated_row.update({
                    'license_plate_bbox_score': matching_record.get('license_plate_bbox_score', '0'),
                    'license_number': matching_record.get('license_number', '0'),
                    'license_number_score': matching_record.get('license_number_score', '0')
                })

            interpolated_output.append(interpolated_row)

    return interpolated_output

with open('/content/test_results.csv', 'r') as csvfile:
    csv_reader = csv.DictReader(csvfile)
    input_data = list(csv_reader)

final_data = interpolate_missing_boxes(input_data)

output_fields = [
    'frame_nmr', 'car_id', 'car_bbox', 'license_plate_bbox',
    'license_plate_bbox_score', 'license_number', 'license_number_score'
]

with open('test_interpolated_results.csv', 'w', newline='') as out_csv:
    csv_writer = csv.DictWriter(out_csv, fieldnames=output_fields)
    csv_writer.writeheader()
    csv_writer.writerows(final_data)
```

```python
# showing output

import ast
import cv2
import numpy as np
import pandas as pd

def draw_stylized_border(image, top_left, bottom_right, color=(0, 255, 0), thickness=10, length_x=200, length_y=200):
    x1, y1 = top_left
    x2, y2 = bottom_right

    cv2.line(image, (x1, y1), (x1, y1 + length_y), color, thickness)
    cv2.line(image, (x1, y1), (x1 + length_x, y1), color, thickness)
    cv2.line(image, (x1, y2), (x1, y2 - length_y), color, thickness)
    cv2.line(image, (x1, y2), (x1 + length_x, y2), color, thickness)
    cv2.line(image, (x2, y1), (x2 - length_x, y1), color, thickness)
    cv2.line(image, (x2, y1), (x2, y1 + length_y), color, thickness)
    cv2.line(image, (x2, y2), (x2, y2 - length_y), color, thickness)
    cv2.line(image, (x2, y2), (x2 - length_x, y2), color, thickness)

    return image
annotations = pd.read_csv('./test_interpolated_results.csv')

input_video_path = 'cars.mp4'
video_capture = cv2.VideoCapture(input_video_path)

fourcc_codec = cv2.VideoWriter_fourcc(*'mp4v')
fps = video_capture.get(cv2.CAP_PROP_FPS)
frame_width = int(video_capture.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(video_capture.get(cv2.CAP_PROP_FRAME_HEIGHT))

output_writer = cv2.VideoWriter('cars_output.mp4', fourcc_codec, fps, (frame_width, frame_height))
plate_info = {}
for vehicle_id in np.unique(annotations['car_id']):
    best_score = np.max(annotations[annotations['car_id'] == vehicle_id]['license_number_score'])


    plate_info[vehicle_id] = {
        'plate_image': None,
        'plate_number': annotations[
            (annotations['car_id'] == vehicle_id) &
            (annotations['license_number_score'] == best_score)
        ]['license_number'].iloc[0]
    }

    best_frame = annotations[
        (annotations['car_id'] == vehicle_id) &
        (annotations['license_number_score'] == best_score)
    ]['frame_nmr'].iloc[0]

    video_capture.set(cv2.CAP_PROP_POS_FRAMES, best_frame)
    success, selected_frame = video_capture.read()

    bbox_string = annotations[
        (annotations['car_id'] == vehicle_id) &
        (annotations['license_number_score'] == best_score)
    ]['license_plate_bbox'].iloc[0]

    x1, y1, x2, y2 = ast.literal_eval(bbox_string.replace('[ ', '[').replace('   ', ' ').replace('  ', ' ').replace(' ', ','))

    crop_img = selected_frame[int(y1):int(y2), int(x1):int(x2), :]
    crop_img = cv2.resize(crop_img, (int((x2 - x1) * 400 / (y2 - y1)), 400))

    plate_info[vehicle_id]['plate_image'] = crop_img

video_capture.set(cv2.CAP_PROP_POS_FRAMES, 0)
frame_index = -1
ret = True

while ret:
    ret, current_frame = video_capture.read()
    frame_index += 1

    if not ret:
        break

    current_annotations = annotations[annotations['frame_nmr'] == frame_index]

    for idx in range(len(current_annotations)):
        row = current_annotations.iloc[idx]
        vehicle_id = row['car_id']

        car_coords = ast.literal_eval(row['car_bbox'].replace('[ ', '[').replace('   ', ' ').replace('  ', ' ').replace(' ', ','))
        draw_stylized_border(
            current_frame,
            (int(car_coords[0]), int(car_coords[1])),
            (int(car_coords[2]), int(car_coords[3])),
            color=(0, 255, 0),
            thickness=25,
            length_x=200,
            length_y=200
        )

        plate_coords = ast.literal_eval(row['license_plate_bbox'].replace('[ ', '[').replace('   ', ' ').replace('  ', ' ').replace(' ', ','))
        cv2.rectangle(
            current_frame,
            (int(plate_coords[0]), int(plate_coords[1])),
            (int(plate_coords[2]), int(plate_coords[3])),
            (0, 0, 255),
            12
        )

        try:
            license_img = plate_info[vehicle_id]['plate_image']
            plate_text = plate_info[vehicle_id]['plate_number']
            h, w, _ = license_img.shape

            x_center = int((car_coords[2] + car_coords[0]) / 2)
            y_top = int(car_coords[1] - h - 100)

            current_frame[y_top:y_top + h,
                          x_center - w // 2:x_center + w // 2, :] = license_img

            current_frame[y_top - 300:y_top,
                          x_center - w // 2:x_center + w // 2, :] = (255, 255, 255)

            (text_w, text_h), _ = cv2.getTextSize(plate_text, cv2.FONT_HERSHEY_SIMPLEX, 4.3, 17)
            cv2.putText(
                current_frame,
                plate_text,
                (x_center - text_w // 2, y_top - 150 + text_h // 2),
                cv2.FONT_HERSHEY_SIMPLEX,
                4.3,
                (0, 0, 0),
                17
            )

        except Exception:
            pass

    output_writer.write(current_frame)

# Cleanup
video_capture.release()
output_writer.release()
```
